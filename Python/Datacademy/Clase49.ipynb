```python
import matplotlib.pyplot as plt
#Se debe añadir el .pyplot con el fin de evitar bugs a la hora de ejecutar funciones de matplotlib
```

## Creación de gráficas con Matplotlib

```python
x = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
y = [1, 2, 5, 10, 21, 40, 54]
```

### Creación de gráfica de lineas

```python
plt.plot(x, y) #Con .plot creamos una grafica de linea con la variables 'x' y 'y'
plt.show() #Con .show se muestra la grafica recién creada
```

```python
plt.plot(x, y, marker='o', linestyle='--', color='c')
#marker genera puntos que permiten ver de manera mas sencilla donde se encuentra cada punto de dato
#linestyle cambia el como se ve la linea que conecta cada punto
#color permite modificar el color de la linea y los puntos en la grafica
plt.show()
```

```python
plt.plot(x, y, marker='o', linestyle='--', color='c')
plt.xlabel('Años')
plt.ylabel('Ganancia')
#Los .xlabel y .ylabel permiten añadir un titulo a cada eje
plt.title('Ganancias por año')
#El .title permite añadir un titulo general a la gráfica
plt.show()
```

### Creación de gráfica de barras

```python
w = ['Data Science', 'Web Development', 'Mobile Development']
v = [400, 500, 150]
#Se crean los datos

plt.bar(w, v)
#Con .bar creamos una grafica de barras con los datos seleccionados
plt.show()
```

### Creación de Histograma

```python
precios = [1000, 2300, 3001, 3450, 4200, 4780, 5100, 5500, 6000, 7500]
contenedores = [i for i in range(0, 8001, 500)]

#los contenedores son la cantidad de bloques en los cuales se separa la información, en este caso están hechos con un for pero pueden ser introducidos individualmente por rangos

plt.hist(precios, contenedores)
#.hist crea el histograma
plt.show()
#.show permite que se muestre el mismo
```

### Creación de gráfica de dispersión - (Scatter Plot)

```python
x = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
y = [1, 2, 5, 10, 21, 40, 54]

#Las graficas de dispersión permiten evaluar si dos variables tienen relación entre ellas

plt.scatter(x, y)
#Con .scatter se crea la gráfica de dispersión

plt.show()
#Con .show se muestra la gráfica
```
